# Задача.
Определить минимально допустимое количество операторов call-центра, если известно, что в среднем, поступает 1,6 звонка в
минуту, средняя продолжительность разговора составляет 88 секунд и число необслуженных клиентов может составить не более 10 %.

```python
# начальные параметры
lam = 1.6 
mu = 1*60/88 
p = lam/mu 

print(f'Интенсивность поступления звонков (мин^(-1)) {lam}')
print(f'Время обработки одной заявки (мин^(-1)) {mu}')
print(f'Коэффициент загруженности СМО {p}')
```

# Вероятность отказа определяется по формуле:


# Относительная пропускная способность


# Абсолютная пропускная способность


```python
n = 1

p0_1 = 1 /(1 + p**n/n)
p_otk_1 = p**n/n * p0_1
Q1 = 1 - p_otk_1
A1 = lam * Q1

if p_otk_1 > 0.1:
    n += 1
else:
    pass

    
print(f'Вероятность = {p0_1}')
print(f'Вероятность отказа = {p_otk_1}')
print(f'Относительная пропускная способность 1 = {Q1}')
print(f'Абсолютная пропускная способность 1 = {A1}')
print(f'Количество операторов = {n}')
```

```python
#n = 2

p0_2 = 1/(1 + p**(n-1)/(n-1) + p**n/(n*(n-1)))
p_otk_2 = p**n/(n*(n-1)) * p0_2
Q2 = 1 - p_otk_2
A2 = lam * Q2

if p_otk_2 > 0.1:
    n += 1
else:
    pass
    
print(f'Вероятность = {p0_2}')
print(f'Вероятность отказа = {p_otk_2}')
print(f'Относительная пропускная способность 1 = {Q2}')
print(f'Абсолютная пропускная способность 1 = {A2}')
print(f'Количество операторов = {n}')
```

```python
#n = 3

p0_3 = 1/(1 + p**(n-2)/(n-2) + p**(n-1)/((n-1)*(n-2)) + p**n/(n*(n-1)*(n-2)))
p_otk_3 = p**n/(n*(n-1)*(n-2)) * p0_3
Q3 = 1 - p_otk_3
A3 = lam * Q3

if p_otk_3 > 0.1:
    n += 1
else:
    pass
    
print(f'Вероятность = {p0_3}')
print(f'Вероятность отказа = {p_otk_3}')
print(f'Относительная пропускная способность 1 = {Q3}')
print(f'Абсолютная пропускная способность 1 = {A3}')
print(f'Количество операторов = {n}')
```

```python
#n = 4

p0_4 = 1/(1 + p**(n-3)/(n-3) + p**(n-2)/((n-2)*(n-3)) + p**(n-1)/((n-1)*(n-2)*(n-3)) + p**n/(n*(n-1)*(n-2)*(n-3)))
p_otk_4 = p**n/(n*(n-1)*(n-2)*(n-3)) * p0_4
Q4 = 1 - p_otk_4
A4 = lam * Q4

if p_otk_4 > 0.1:
    n += 1
else:
    pass
    
print(f'Вероятность = {p0_4}')
print(f'Вероятность отказа = {p_otk_4}')
print(f'Относительная пропускная способность 1 = {Q4}')
print(f'Абсолютная пропускная способность 1 = {A4}')
print(f'Количество операторов = {n}')
```

```python
#n = 5

p0_5 = 1/(1 + p**(n-4)/(n-4) + p**(n-3)/((n-3)*(n-4)) + p**(n-2)/((n-2)*(n-3)*(n-4)) + p**(n-1)/((n-1)*(n-2)*(n-3)*(n-4)) + p**n/(n*(n-1)*(n-2)*(n-3)*(n-4)))
p_otk_5 = p**n/(n*(n-1)*(n-2)*(n-3)*(n-4)) * p0_5
Q5 = 1 - p_otk_5
A5 = lam * Q5

if p_otk_5 > 0.1:
    n += 1
else:
    pass
    
print(f'Вероятность = {p0_5}')
print(f'Вероятность отказа = {p_otk_5}')
print(f'Относительная пропускная способность 1 = {Q5}')
print(f'Абсолютная пропускная способность 1 = {A5}')
print(f'Количество операторов = {n}')
```
